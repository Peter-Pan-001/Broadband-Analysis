## Aggeregate

```python
bytes_sec_mean = np.zeros(shape = (14, ))
bytes_sec_std = np.zeros(shape = (14, ))
availability = np.zeros(shape = (14, ))

# calculate column bytes_sec_mean
bytes_sec_mean[0] = curr_httpget.loc[curr_httpget['bytes_sec'] != 0, 'bytes_sec'].mean()
bytes_sec_mean[1] = curr_httpgetmt.loc[(curr_httpgetmt['bytes_sec'] != 0) & (curr_httpgetmt['threads'] == 3),'bytes_sec'].mean()
bytes_sec_mean[2] = curr_httpgetmt.loc[(curr_httpgetmt['bytes_sec'] != 0) & (curr_httpgetmt['threads'] == 8),'bytes_sec'].mean()
bytes_sec_mean[3] = curr_httpgetmt.loc[(curr_httpgetmt['bytes_sec'] != 0) & (curr_httpgetmt['threads'] == 12),'bytes_sec'].mean()
bytes_sec_mean[4] = curr_httppost.loc[curr_httppost['bytes_sec'] != 0, 'bytes_sec'].mean()
bytes_sec_mean[5] = curr_httppostmt.loc[(curr_httppostmt['bytes_sec'] != 0) & (curr_httppostmt['threads'] == 3),'bytes_sec'].mean()
bytes_sec_mean[6] = curr_httppostmt.loc[(curr_httppostmt['bytes_sec'] != 0) & (curr_httppostmt['threads'] == 8),'bytes_sec'].mean()
bytes_sec_mean[7] = curr_httppostmt.loc[(curr_httppostmt['bytes_sec'] != 0) & (curr_httppostmt['threads'] == 12),'bytes_sec'].mean()
#bytes_sec_mean[8] = curr_httpget6.loc[curr_httpget6['bytes_sec'] != 0, 'bytes_sec'].mean()
bytes_sec_mean[9] = curr_httpgetmt6.loc[(curr_httpgetmt6['bytes_sec'] != 0) & (curr_httpgetmt6['threads'] == 3),'bytes_sec'].mean()
bytes_sec_mean[10] = curr_httpgetmt6.loc[(curr_httpgetmt6['bytes_sec'] != 0) & (curr_httpgetmt6['threads'] == 8),'bytes_sec'].mean()
#bytes_sec_mean[11] = curr_httppost6.loc[curr_httppost6['bytes_sec'] != 0, 'bytes_sec'].mean()
bytes_sec_mean[12] = curr_httppostmt6.loc[(curr_httppostmt6['bytes_sec'] != 0) & (curr_httppostmt6['threads'] == 3),'bytes_sec'].mean()
bytes_sec_mean[13] = curr_httppostmt6.loc[(curr_httppostmt6['bytes_sec'] != 0) & (curr_httppostmt6['threads'] == 8),'bytes_sec'].mean()


# calculate column bytes_sec_std
bytes_sec_std[0] = curr_httpget.loc[curr_httpget['bytes_sec'] != 0, 'bytes_sec'].std()
bytes_sec_std[1] = curr_httpgetmt.loc[(curr_httpgetmt['bytes_sec'] != 0) & (curr_httpgetmt['threads'] == 3),'bytes_sec'].std()
bytes_sec_std[2] = curr_httpgetmt.loc[(curr_httpgetmt['bytes_sec'] != 0) & (curr_httpgetmt['threads'] == 8),'bytes_sec'].std()
bytes_sec_std[3] = curr_httpgetmt.loc[(curr_httpgetmt['bytes_sec'] != 0) & (curr_httpgetmt['threads'] == 12),'bytes_sec'].std()
bytes_sec_std[4] = curr_httppost.loc[curr_httppost['bytes_sec'] != 0, 'bytes_sec'].std()
bytes_sec_std[5] = curr_httppostmt.loc[(curr_httppostmt['bytes_sec'] != 0) & (curr_httppostmt['threads'] == 3),'bytes_sec'].std()
bytes_sec_std[6] = curr_httppostmt.loc[(curr_httppostmt['bytes_sec'] != 0) & (curr_httppostmt['threads'] == 8),'bytes_sec'].std()
bytes_sec_std[7] = curr_httppostmt.loc[(curr_httppostmt['bytes_sec'] != 0) & (curr_httppostmt['threads'] == 12),'bytes_sec'].std()
#bytes_sec_std[8] = curr_httpget6.loc[curr_httpget6['bytes_sec'] != 0, 'bytes_sec'].std()
bytes_sec_std[9] = curr_httpgetmt6.loc[(curr_httpgetmt6['bytes_sec'] != 0) & (curr_httpgetmt6['threads'] == 3),'bytes_sec'].std()
bytes_sec_std[10] = curr_httpgetmt6.loc[(curr_httpgetmt6['bytes_sec'] != 0) & (curr_httpgetmt6['threads'] == 8),'bytes_sec'].std()
#bytes_sec_std[11] = curr_httppost6.loc[curr_httppost6['bytes_sec'] != 0, 'bytes_sec'].std()
bytes_sec_std[12] = curr_httppostmt6.loc[(curr_httppostmt6['bytes_sec'] != 0) & (curr_httppostmt6['threads'] == 3),'bytes_sec'].std()
bytes_sec_std[13] = curr_httppostmt6.loc[(curr_httppostmt6['bytes_sec'] != 0) & (curr_httppostmt6['threads'] == 8),'bytes_sec'].std()

# calculate availability
availability[0] = np.sum(np.array(curr_httpget[['successes']])) / np.sum(np.array(curr_httpget[['successes','failures']]))
availability[1] = np.sum(np.array(curr_httpgetmt.loc[curr_httpgetmt['threads'] == 3, ['successes']])) / \
                  np.sum(np.array(curr_httpgetmt.loc[curr_httpgetmt['threads'] == 3, ['successes','failures']]))
availability[2] = np.sum(np.array(curr_httpgetmt.loc[curr_httpgetmt['threads'] == 8, ['successes']])) / \
                  np.sum(np.array(curr_httpgetmt.loc[curr_httpgetmt['threads'] == 8, ['successes','failures']]))
availability[3] = np.sum(np.array(curr_httpgetmt.loc[curr_httpgetmt['threads'] == 12, ['successes']])) / \
                  np.sum(np.array(curr_httpgetmt.loc[curr_httpgetmt['threads'] == 12, ['successes','failures']]))
availability[4] = np.sum(np.array(curr_httppost[['successes']])) / \
                  np.sum(np.array(curr_httppost[['successes','failures']]))
availability[5] = np.sum(np.array(curr_httppostmt.loc[curr_httppostmt['threads'] == 3, ['successes']])) / \
                  np.sum(np.array(curr_httppostmt.loc[curr_httppostmt['threads'] == 3, ['successes','failures']]))
availability[6] = np.sum(np.array(curr_httppostmt.loc[curr_httppostmt['threads'] == 8, ['successes']])) / \
                  np.sum(np.array(curr_httppostmt.loc[curr_httppostmt['threads'] == 8, ['successes','failures']]))
availability[7] = np.sum(np.array(curr_httppostmt.loc[curr_httppostmt['threads'] == 12, ['successes']])) / \
                  np.sum(np.array(curr_httppostmt.loc[curr_httppostmt['threads'] == 12, ['successes','failures']]))
#availability[8] = np.sum(np.array(curr_httpget6[['successes']])) / np.sum(np.array(curr_httpget6[['successes','failures']]))
availability[9] = np.sum(np.array(curr_httpgetmt6.loc[curr_httpgetmt6['threads'] == 3, ['successes']])) / \
                  np.sum(np.array(curr_httpgetmt6.loc[curr_httpgetmt6['threads'] == 3, ['successes','failures']]))
availability[10] = np.sum(np.array(curr_httpgetmt6.loc[curr_httpgetmt6['threads'] == 8, ['successes']])) / \
                  np.sum(np.array(curr_httpgetmt6.loc[curr_httpgetmt6['threads'] == 8, ['successes','failures']]))
#availability[11] = np.sum(np.array(curr_httppost6[['successes']])) / np.sum(np.array(curr_httppost6[['successes','failures']]))
availability[12] = np.sum(np.array(curr_httppostmt6.loc[curr_httppostmt6['threads'] == 3, ['successes']])) / \
                  np.sum(np.array(curr_httppostmt6.loc[curr_httppostmt6['threads'] == 3, ['successes','failures']]))
availability[13] = np.sum(np.array(curr_httppostmt6.loc[curr_httppostmt6['threads'] == 8, ['successes']])) / \
                  np.sum(np.array(curr_httppostmt6.loc[curr_httppostmt6['threads'] == 8, ['successes','failures']]))

# create aggeragate matrix
final_matrix_1 = []
list1 = ['2018-01','IPV4', 'Down', '1', bytes_sec_mean[0], bytes_sec_std[0], availability[0]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV4', 'Down', '3', bytes_sec_mean[1], bytes_sec_std[1], availability[1]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV4', 'Down', '8', bytes_sec_mean[2], bytes_sec_std[2], availability[2]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV4', 'Down', '12', bytes_sec_mean[3], bytes_sec_std[3], availability[3]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV4', 'Up', '1', bytes_sec_mean[4], bytes_sec_std[4], availability[4]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV4', 'Up', '3', bytes_sec_mean[5], bytes_sec_std[5], availability[5]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV4', 'Up', '8', bytes_sec_mean[6], bytes_sec_std[6], availability[6]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV4', 'Up', '12', bytes_sec_mean[7], bytes_sec_std[7], availability[7]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV6', 'Down', '1', bytes_sec_mean[8], bytes_sec_std[8], availability[8]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV6', 'Down', '3', bytes_sec_mean[9], bytes_sec_std[9], availability[9]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV6', 'Down', '8', bytes_sec_mean[10], bytes_sec_std[10], availability[10]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV6', 'Up', '1', bytes_sec_mean[11], bytes_sec_std[11], availability[11]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV6', 'Up', '3', bytes_sec_mean[12], bytes_sec_std[12], availability[12]]
final_matrix_1.append(list1)
list1 = ['2018-01','IPV6', 'Up', '8', bytes_sec_mean[13], bytes_sec_std[13], availability[13]]
final_matrix_1.append(list1)


# matrix to dataframe
df_final_1 = pd.DataFrame(final_matrix_1, columns = ['Date', 'IPV4/IPV6', 'Up/Down', 'Threads', \
                                                     'Throughout_mean', 'Throughout_std', 'Availability'])

df_final_1
```

```python
# Standardization
df_final_1['Throughout_mean'] = (df_final_1['Throughout_mean'] - df_final_1['Throughout_mean'].mean()) / \
                                (df_final_1['Throughout_mean'].std())
df_final_1['Throughout_std'] = (df_final_1['Throughout_std'] - df_final_1['Throughout_std'].mean()) / \
                               (df_final_1['Throughout_std'].std())
df_final_1
```

## Geography

```python
geo = pd.read_excel('Unit-Profile-sept2017.xlsx')
geo.head()
```

```python
geo = geo[['unit_id', 'ISP', 'Technology', 'State', 'Census']]
geo.head()
```

```python
curr_httpget_g = curr_httpget[['unit_id', 'bytes_sec', 'successes', 'failures']]
curr_httpget_g = curr_httpget_g.groupby('unit_id').mean()
curr_httpget_g.reset_index(['unit_id'], inplace = True)
curr_httpget_g.head()
```

```python
httpget_merge = curr_httpget_g.merge(geo, how = 'inner', left_on = 'unit_id', right_on = 'unit_id')
httpget_merge.head()
```

```python
httpget_merge['Availability'] = httpget_merge['successes'] / \
                                (httpget_merge['successes'] + httpget_merge['failures'])
httpget_merge['ISP-Technology'] = httpget_merge[['ISP', 'Technology']].apply(lambda x: '-'.join(x), axis=1)
httpget_merge.head()
```

#### Group by ISP and Technology

```python
fig, ax = plt.subplots(figsize=(20,20))
order = httpget_merge.groupby('ISP-Technology')["bytes_sec"].mean().sort_values().iloc[::-1].index
_ = sns.boxplot(x = 'bytes_sec',y = 'ISP-Technology', data=httpget_merge, \
                orient="h", palette="Set1", order = order, ax = ax)
_ = sns.swarmplot(x = 'bytes_sec',y = 'ISP-Technology', data=httpget_merge, \
                orient="h", palette="Set1", order = order, ax = ax)
_ = plt.title('Byte_sec Group By ISP-Technology', fontsize = 20)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
order = httpget_merge.groupby('ISP-Technology')["Availability"].mean().sort_values().iloc[::-1].index
_ = sns.boxplot(x = 'Availability',y = 'ISP-Technology', data=httpget_merge, \
                orient="h", palette="Set1", order = order, ax = ax)
_ = sns.swarmplot(x = 'Availability',y = 'ISP-Technology', data=httpget_merge, \
                orient="h", palette="Set1", order = order, ax = ax)
_ = plt.title('Availability Group By ISP-Technology', fontsize = 20)
```
